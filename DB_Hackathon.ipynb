```python
gradsBoard = {}
teamsBoard = {}
```

```python
def find_team_index(team_name):
    for key,value in teamsBoard.items():
        if team_name == value.name:
            return key
    print("could not find a team with the specified team name")
    return None
    
def find_grad_index(grad_name):
    for key,value in gradsBoard.items():
        if grad_name == value.name:
            return key
    print("could not find a grad with the specified grad name")
    return None
```

```python
class Grad:
    matched = False
    team = None
    team_name = ""
    
    def __init__(self, name, preferences, id = None, skills = ""):
        self.name = name
        self.id = id
        self.skills = skills
        self.preferences = preferences
        
    def make_proposition(self, team_name):
        key = find_team_index(team_name)
        if key != None:
            team = teamsBoard[key]
            proposition = team.test_proposition(self.name)
            
            if proposition == True:
                self.update_team(team)
        pass
    
    def update_team(self, team):
        # tell my current team I am no longer there
        if self.matched == True:
            self.team.matched = False
            self.team.grad = None
            self.team.grad_name = ""
        
        # tell my new team I am one of theirs
        self.matched = True
        self.team = team
        self.team_name = team.name
        
        pass
    
    pass
```

```python
class Team:
    matched = False
    grad = None
    grad_name = ""
    
    def __init__(self, name, preferences, id = None, skills = ""):
        self.name = name
        self.id = id
        self.preferences = preferences
        self.skills = skills
        
    def test_proposition(self, prop_grad):
        proposed_grad = gradsBoard[find_grad_index(prop_grad)]
        if self.matched == False:
            print("I currently have no grad so I accept")
            print()
            self.update_grad(proposed_grad)
            return True
        else:
            if proposed_grad.name in self.preferences:
                if self.grad_name not in self.preferences:
                    print("I prefer this grad over what I currently have so I accept their proposition")
                    self.update_grad(proposed_grad)
                    return True
                else:
                    if self.preferences.index(proposed_grad.name) < self.preferences.index(self.grad_name):
                        print("I prefer this grad over what I currently have so I accept their proposition")
                        print()
                        self.update_grad(proposed_grad)
                        return True
                    print("I have no preference for this grad over what I currently have so I reject")
                    print()
                    return False
            else:
                print("I have no preference for this grad over what I currently have so I reject")
                print()
                
                return False
        
    
    def update_grad(self, grad):
#         I also need to tell the grad I currently have that he is no longer matched
        if self.matched == True:
            self.grad.matched = False
            self.grad.team = None
            self.grad.team_name = ""
        
        self.matched = True
        self.grad = grad
        self.grad_name = grad.name
        pass
    
    pass
```

```python
for index, row in grad_df.iterrows():
    gradsBoard[index] = Grad(row[0], [row[1], row[2], row[3], row[4], row[5]], id = index, skills = [row[6], row[7], row[8]])

for index, row in manager_df.iterrows():
    teamsBoard[index] = Team(row[0], [row[1], row[2], row[3], row[4], row[5]], id = index, skills = [row[6], row[7], row[8]])
```
